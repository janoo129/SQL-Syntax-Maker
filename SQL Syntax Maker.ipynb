```python
#Creating a class so that it can be assigned to any object and used for this program.
class SqlSyntaxMaker:     
    
    def update(self):#this function will return the update command syntax in SQL.
        
        #first we need to know the table name.
        table_name = input("What is table name ? : ")
        
        #column which we need to update.
        column_name = input("Mention column name you want to change values in : ")
        
        #values which we need to update in above mentioned column.
        values = input(f"Mention values you want to update in column {column_name} : ")
        
        #condition as per which values will be updated.
        condition = input("Mention condition where you want to change the values : ")
        
        #here screen will be cleared and only answer will be visible.
        clear_output()
        
        #following is the answer for update function.
        print("Answer :\n")
        print(f"UPDATE {table_name}\n SET {column_name} = {values} \n WHERE {condition}; ")
    
    def delete(self):#this function will return the delete command syntax in SQl.
        
        #first we need to know the table name.
        table_name = input("What is table name ? : ")
        
        #condition as per which rows will be deleted.
        condition = input(f"What is the condition as per which you want to delete rows from table {table_name} ? : ")
        
        #here screen will be cleared and only answer will be visible.
        clear_output()
        
        #following is the answer for delete function.
        print("\nAnswer :")
        print(f"DELETE FROM {table_name}\n WHERE {condition};")
        
        
    def joins(self):#this will return different joins syntax in SQl.
        
        #we need to know the first table name.
        table_name = input("What is table name ? : ")
        
        #second table name.
        sec_table_name = input("What is other table's name ? : ")
        
        #here we are asking which join we need to perform.
        print("\nWhich type of join you want to apply for these tables ?\n A) INNER JOIN\n B) LEFT JOIN\n C) RIGHT JOIN\n D) FULL OUTER JOIN")
        
        #option selection as per which syntax will be made.
        sel = input("Mention your answer in form of A, B, C or D : ")
        
        #common column in both tables on which join will be performed.
        column = input("Which column you want to join from both tables ? : ")
        
        #if condition to determine join type as per the user's selection.
        if sel == 'a':
            option = "INNER JOIN"
        elif sel == 'b':
            option = "LEFT JOIN"
        elif sel == 'c':
            option = "RIGHT JOIN"
        elif sel == 'd':
            option = "FULL OUTR JOIN"

        #here screen will be cleared and only answer will be visible.
        clear_output()
        
        #following is the answer for joins function.
        print("Answer :\n")
        print(f"SELECT * FROM {table_name}\n {option} {sec_table_name} ON {table_name}.{column} = {sec_table_name}.{column};")
    
    
    def insert(self):#this will return the insert command syntax in SQl.
        
        
        table_name = input("What is table name ?: ")
        
        #we are asking column names here in one input in order to avoid complexity.
        columns = input("Please mention column names separated by ',' comma : ")
        
        #number of rows to run "for loop" so that only these number of rows will be added in the syntax.
        rows = int(input(f"How many rows you want to insert into {table_name}"))
        
        #variable to collect results from the below mentioned "for loop"
        container = []
        
        #here screen will be cleared.
        clear_output()
        
        #we are informing that program will ask to insert values as many times as user has mentioned in rows tab.
        print(f"There are {rows} rows hence you will have to type values {rows} times")
        
        #this for loop will take the values and add to container variable in the form of list.
        for val in range(rows):               
            values = input("Please mention values separated by ',' comma : ")
            container.append(values)
        
        #here screen will be cleared and only answer will be visible.
        clear_output()
        
        #here is the solution to insert function.
        print(f"INSERT INTO {table_name} ({columns})\n VALUES")

        last = 0 #this variable is created just to make sure that last row will not end with ',' but ';' instead.
        
        for row in container:
            last += 1

            if last == rows:
                print(" ("+row+");")
            else:
                print(" ("+row+"),")
            
               
    def drop(self):#this function will return the drop command syntax in SQl.
        
        #first we are hinting that there is another option if only data needs to be deleted from the table and not the table itself.
        print("Please note that DROP command will delete entire table,")
        print("if you just want to delete all data from table then try Truncate.")
        print("\nWhat do you want to do ? \nA) DROP\nB) TRUNCATE\n")
        
        #as per the selection syntax will be updated as drop or truncate.
        selection = input("Please mention your option here in form of A or B: ").lower()
        
        #here we are asking table name.
        table_name = input("What is table name ? : ")
        
        #here screen will be cleared and only answer will be visible.
        clear_output()
        
        #here is the solution for drop function.
        print("Answer :\n")
        if selection == 'a':
            print(f"DROP TABLE {table_name};")
        elif selection == 'b':
            print(f"TRUNCATE TABLE {table_name};")
        else:
            print("Please mention correct option !")
            
            
    def alter(self): #this will return alter command syntax in SQl.
            
            #as alter command modifies many parts of the table hence we will ask what needs to be altered.
            print("What do you want to alter ?\n")
            print("A) Alter Table Name\nB) Alter Column name\nC) Alter Column Data Type\nD) Add Column\nE) Drop Column")      
            alter_type = input("Please write your option in form of 'A', 'B', 'C', 'D' or 'E':").lower()
            
            #here screen will be cleared and only answer will be visible.
            clear_output()
            
            #according to selection alter command will be performed and solution will be returned.
            if alter_type == 'a':
                current_name = input("What is your current table name ? :")
                new_name = input("What is new table name ? :")
                print("\nAnswer :")
                print(f"ALTER TABLE {current_name} \n\tRENAME TO {new_name};")
            elif alter_type == 'b':
                table_name = input("What is table name ? :")
                column_name = input("What is current column name ? :")
                new_column_name = input("What is new column name ? :")
                print("\nAnswer:")
                print(f"ALTER TABLE {table_name} \n\tRENAME COLUMN {column_name} to {new_column_name};")
            elif alter_type == 'c':
                table_name = input("What is table name ?  :")
                column_name = input("What is column name ?:")
                dtype = input("What is data type ? :")
                print("\nAnswer:")
                print(f"ALTER TABLE {table_name} \n\tMODIFY COLUMN {column_name} {dtype};")
            elif alter_type == 'd':
                table_name = input("What is table name ? : ")
                new_column_name = input("What is new column name ? : ")
                dtype = input(f"What is {new_column_name}'s data type ? : ")
                print("\nAnswer:")
                print(f"ALTER TABLE {table_name} \n\tADD {new_column_name} {dtype};")
            elif alter_type == 'e':
                table_name = input("What is table name ? : ")
                column_name = input("What is column name ? : ")
                print("\nAnswer:")
                print(f"ALTER TABLE {table_name} \n\tDROP COLUMN {column_name};")
            else:
                print("Please input correct option.")
                alter()
                
    
    def create(self):#this will return create command syntax in SQl.
        
        
        new_table_name = input("What is new table name ? : ")
        
        #no. of columns to determine how many times program will ask for column name and it's data type.
        columns = int(input(f"How many columns table {new_table_name} have ?"))
        
        #empty lists which will be used later.
        more_columns = []
        more_dtypes = []
        
        #this is first column and as it asks 1st with 'st', hence written separately before rest of the program.
        Column1 = input("What is 1st Column's name ? : ")
        Column1_dtype = input(f"What is {Column1}'s data type ? : ")
        
        #adding the result to empty lists mentioned above.
        more_columns.append(Column1)
        more_dtypes.append(Column1_dtype)
        
        
        # if columns are more than 1 then it will ask 2nd column name and as it has 'nd', it has been written separately.
        if columns > 1:
            Column2 = input("What is 2nd Column's name ? : ") 
            Column2_dtype = input(f"What is {Column2}'s data type ? : ")
            
            #adding the result to lists mentioned above which has 1st column details already.
            more_columns.append(Column2)
            more_dtypes.append(Column2_dtype)
            
            
            # if columns are more than 2 then it will ask 3rd column name and as it has 'rd', it has been written separately.
            if columns > 2:
                Column3 = input("What is 3rd Column's name ? : ")
                Column3_dtype = input(f"What is {Column3}'s data type ? : ")
                
                #adding the result to lists mentioned above which has 1st column and 2nd column details already.
                more_columns.append(Column3)
                more_dtypes.append(Column3_dtype)                   
                    
                if columns > 3:
                    col_no = 3                               
                    
                    #here as most of columns numbers will end with 'th' like 4th,5th,6th etc. hence running for loop to iterate over.
                    for col in range(columns-3): #we are subtracting 3 as we have already added 3 columns above.
                        
                        #this variable will return numbers which we need to ask like 4th,5th,6th etc.
                        col_no += 1
                        Add_Column = input(f"What is {col_no}th Column's name ? : ")
                        
                        #adding the result to lists mentioned above.
                        more_columns.append(Add_Column)                        
                        
                        Add_dtypes = input(f"What is {Add_Column}'s data type ? : ")
                        
                        #adding the result to lists mentioned above.
                        more_dtypes.append(Add_dtypes)
        
        #here screen will be cleared and only answer will be visible.
        clear_output()
        
        #solution for create funtion.
        print("Answer :")
        print(f"\nCREATE TABLE {new_table_name} (")
        
        #zipping 2 different lists which will be unpacked later.
        Col_tuples = zip(more_columns,more_dtypes)
        
        #this variable will make sure that only 'for loop' runs only number of times, rows are mentioned for the table.
        last = 0
        
        #tuple unpacking and then to be printed in the solution.
        for col,dtypes in Col_tuples:
            last += 1
            if last == columns:
                print(f"\n\t{col} {dtypes}")
            else:
                print(f"\n\t{col} {dtypes},")
        print("\t);")                            
        
        
```

```python
#to run the while loop setting this variable to True.
code = True

#assigning sqlsyntaxmaker class to this object.
syntax = SqlSyntaxMaker

#Welcome messages
print("Welcome to SQL Syntax Maker\n\nWhat do you want to do today ?\n")
print("Please select your option from below:")
print("A) CREATE\nB) INSERT\nC) UPDATE\nD) DELETE\nE) DROP\nF) ALTER\nG) JOINS")

#while loop to continue this program until user selects no.
while code:   
    
    sel = input("Mention your option in form of A, B, C : ").lower()
     
    # functions will be returned as per the selection.
    if sel == 'a':
        clear_output()
        print("You are going to create a table now.\n")
        syntax.create(SqlSyntaxMaker)
    elif sel == 'b':
        clear_output()
        print("You are going to insert values into a table now.\n")
        syntax.insert(SqlSyntaxMaker)
    elif sel == 'c':
        clear_output()
        print("You are going to update a table now.\n")
        syntax.update(SqlSyntaxMaker)
    elif sel == 'd':
        clear_output()
        print("You are going to delete values from a table now.\n")
        syntax.delete(SqlSyntaxMaker)
    elif sel == 'e':
        clear_output()
        print("You are going to delete a table now.\n")
        syntax.drop(SqlSyntaxMaker)
    elif sel == 'f':
        clear_output()
        print("You are going to alter a table now.\n")
        syntax.alter(SqlSyntaxMaker)
    elif sel == 'g':
        clear_output()
        print("You are going to use joins for a table now.\n")
        syntax.joins(SqlSyntaxMaker)
    else:
        clear_output()
        print("Please mention correct option.")   
    
    open = True
    
    while open:
        on = input("Do you want to continue ? Y or N : ").lower()
        if on == 'y':
            clear_output()
            print("Please select your option from below:")
            print("A) CREATE\nB) INSERT\nC) UPDATE\nD) DELETE\nE) DROP\nF) ALTER\nG) JOINS")
            open = False
        elif on == 'n':
            clear_output()
            print("Thank you for using SQL Syntax Maker!")
            code = False
            open = False
        else:
            print('Please submit correct option!')
        
        
```
